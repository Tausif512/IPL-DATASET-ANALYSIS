```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
#loading ipl match dataset
ipl=pd.read_csv('matches.csv')
```

```python
ipl.head()
```

```python
ipl.shape
```

```python
#getting the frequency of most man of the match award
ipl['player_of_match'].value_counts()
```

```python
#getting top 10 man of the match
ipl['player_of_match'].value_counts()[0:10]
```

```python
#getting top 5 player
ipl['player_of_match'].value_counts()[0:5]
```

```python
#make list of top 5 player name only
list(ipl['player_of_match'].value_counts()[0:5].keys())

```

```python
#making a bar plot for top 5 man of the match
plt.figure(figsize=(5,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color='g')
plt.show()
```

```python
#frequency of result   (ipl.value_counts('result')) 
ipl['result'].value_counts()
```

```python
#find the no of toss winner w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#making histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title('distribution of runs')
plt.xlabel('runs')
plt.show()
```

```python
#no of winner after batting first
batting_first['winner'].value_counts()
```

```python
list(batting_first['winner'].value_counts()[0:3].keys())
```

```python
#make a bar plot of top 3 teams with most wins after batting_first
plt.figure(figsize=(5,5))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=['red','green','blue'])
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extracting record team win after second match
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
#making histogram win with res to no of wiket
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
#making bar plot for top 3 winner after batting second
plt.figure(figsize=(5,5))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=['purple','blue','red'])
plt.show()
```

```python
#pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
